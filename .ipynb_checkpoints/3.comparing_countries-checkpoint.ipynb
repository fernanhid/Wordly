```python
import pandas as pd
import seaborn as sns
% matplotlib inline
```

```python
australia = pd.read_pickle('australia_month.pkl')
india = pd.read_pickle('india_month.pkl')
america = pd.read_pickle('nyt_month.pkl')
uk = pd.read_pickle('guardian_month.pkl')
peru = pd.read_pickle('republica_month.pkl')
china = pd.read_pickle('china_month.pkl')
```

```python
country_dict = {'australia': australia, 'india': india, 'america': america, 'uk': uk, 'peru': peru, 'china': china}
```

```python
# Function to see which countries talk about a specific subject
def check(topic):
    lst = []
    for country in country_dict.keys():
        if topic not in country_dict[country].columns:
            lst.append(country)
    return lst
```

# Tracking Topics per Month

## Brexit

```python
rc={'xtick.labelsize': 30, 'ytick.labelsize': 30, 'axes.labelsize': 25, 'font.size': 25, 'legend.fontsize': 35.0, 'axes.titlesize': 20, "figure.figsize": [35, 14]}
sns.set(rc = rc)
```

```python
topic= 'Brexit'
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', "Jun", "July", 'August']
raw_brexit = pd.DataFrame({'India': india[topic].values,
                      'U.S.': america[topic].values,'U.K.': uk[topic].values,
                      'Peru': peru[topic].values,'China': china[topic].values}, index = months)

brexit = pd.melt(raw_brexit)
brexit.columns = ['Country', 'NMF Value']
brexit['month'] = months *5
brexit['NMF Value'] = brexit['NMF Value'] *100
sns.pointplot(x = 'month', y =  'NMF Value', hue = 'Country', data = brexit)
plt.title('Brexit', fontsize=40)
plt.xlabel('Months', fontsize=35)
_ = plt.ylabel('% of Articles', fontsize=35)
```

# Turkey Coup

```python
topic= 'Turkey Coup'
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', "Jun", "July"]
raw_turkey_coup = pd.DataFrame({'India': india[topic].values[:7],
                      'U.S.': america[topic].values[:7],'U.K.': uk[topic].values[:7],
                      'Peru': peru[topic].values[:7],'China': china[topic].values[:7],
                               'Australia': australia[topic].values[:7]}, index = months)

turkey_coup = pd.melt(raw_turkey_coup)
turkey_coup['month'] = months *6
sns.pointplot(x = 'month', y =  'value', hue = 'variable', data = turkey_coup)
plt.title('Turkish Coup', fontsize=40)
plt.xlabel('Months', fontsize=35)
_ = plt.ylabel('% of Articles', fontsize=35)
```
